```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
```

```python
# Embedded CSV Data
csv_data = """
Name,Age,Marks,Attendance,StudyHours
Rahul,20,85,90,5
Priya,21,92,95,6
Arjun,19,78,88,4
Sneha,20,95,98,7
Kiran,22,67,80,3
Meena,21,89,92,5
Ravi,20,73,85,4
Anjali,19,91,97,6
Vikram,22,76,84,4
Pooja,20,88,93,5
"""
```

```python
# Load Dataset
df = pd.read_csv(StringIO(csv_data))

print("=" * 50)
print("DATASET PREVIEW")
print("=" * 50)
print(df.head())

print("\n")

print("=" * 50)
print("MISSING VALUE ANALYSIS")
print("=" * 50)
print(df.isnull().sum())

print("\n")

print("=" * 50)
print("STATISTICAL SUMMARY")
print("=" * 50)
print(df.describe())
```

```python
# Visualization 1 - Histogram
plt.figure(figsize=(6, 4))
plt.hist(df["Marks"], bins=5)
plt.title("Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
# Visualization 2 - Scatter Plot
plt.figure(figsize=(6, 4))
plt.scatter(df["StudyHours"], df["Marks"])
plt.title("Study Hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
# Visualization 3 - Box Plot
plt.figure(figsize=(6, 4))
plt.boxplot(df["Attendance"])
plt.title("Box Plot of Attendance")
plt.show()
```
